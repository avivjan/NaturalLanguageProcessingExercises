```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
n = 2
x = np.linspace(0, 10, n)
plt.plot(x, 2*x+3)
plt.xlabel('x')
plt.ylabel('y')
plt.title("y = 2x + 3")
plt.legend()
plt.grid(True)
plt.show()
```

```python
N = 15
x = np.linspace(0, 10, N)
plt.plot(x, x**2, 'o')
plt.xlabel('x')
plt.ylabel('y')
plt.title(" y= x^2")
plt.legend()
plt.grid(True)
plt.show()
```

```python
N = 50
x = np.linspace(0, 10, N, endpoint=True)
epsilon = np.random.normal(0, 1, len(x))
plt.plot(x, x**2 + epsilon, 'o')
plt.xlabel('x')
plt.ylabel('y')
plt.title(" y= x^2 + eps")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the sigmoid function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Generate x values in the range [-10, 10]
x_values = np.linspace(-10, 10, 100)

# Calculate y values using the sigmoid function
sigmoidx_values = sigmoid(x_values)
sigmoid_minusx_values = sigmoid(-1 * x_values)
sigmiod_derivitive = sigmoidx_values * sigmoid_minusx_values

# Plot the sigmoid function
plt.figure(figsize=(8, 6))
plt.plot(x_values, sigmoidx_values, label='Sigmoid(x) Function', color='blue')
plt.plot(x_values, sigmoid_minusx_values, label='Sigmoid(-x) Function', color='green')
plt.plot(x_values, sigmiod_derivitive, label='Sigmoid derivitive Function', color='red')
plt.title('Sigmoid Function')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()
plt.show()

#f(x) + f(-x) = 1
#f'(x) = exp(-x)/(1+exp(x))^2 = f(x) * (1 - f(x)) = f(x) * f(-x)
```
